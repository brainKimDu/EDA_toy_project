##  이제 유의미한 정보를 뽑아내봅시다.
우선 우리는 합격률을 모른다는 가정을 하고 시작을 해봅시다.

```python
del Electricity["필기합격률(%)"]
Electricity.head()
```

필기합격률을 지웠고

```python
del Electricity["실기합격률(%)"]
Electricity.head()
```

실기합격률을 지웠습니다.

```python
Electricity["필기합격률"] = Electricity["필기합격자"] / Electricity["필기응시자"] * 100
Electricity["최종합격률"] = Electricity["실기합격자"] / Electricity["실기응시자"] * 100
Electricity.head()
```

다음의 계산식을 통해 필기합격률과 최종합격률을 구해서 데이터에 추가하였습니다.

```python
Electricity.sort_values(by="최종합격률", ascending=True).head(5)
```

전기기사 최종 합격률(실기)이 가장 낮았던 해는  2009, 2010, 2018, 2012, 2013년 입니다.

```python
Electricity.sort_values(by="최종합격률", ascending=False).head(5)
```

전기기사 최종 합격률이 가장 높았던 해는  2002, 2011, 2004, 2019, 2017년 입니다.

```python
Electricity.sort_values(by="필기응시자", ascending=False).head(5)
```

여기서 신기한 것은 전기기사 필기 응시자들은 2021, 2020, 2019, 2018, 2017년 순으로 응시자는 지속적으로 증가하는 추세입니다.

```python
Electricity.sort_values(by="실기응시자", ascending=False).head(5)
```

응시자가 많아지면서 실기응시자도 최근 5년이 가장 많습니다

```python
Electricity.sort_values(by="실기합격자", ascending=False).head(8)
```

최종합격자의 수도 2018년을 제외하고 최근 5년이 가장 많습니다

## 데이터시각화

```python
import matplotlib.pyplot as plt
get_ipython().run_line_magic("matplotlib", "inline")
```

맨 윗줄이 가장 중요합니다. matplotlib는 데이터를 시각화해주는 라이브러리 입니다.

```python
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

f_path = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

단점은 matplotlib는 한글출력 문제가 있습니다. 그렇기에 해당 코드를 입력해서 한글문제를 해결해야합니다

```python
Electricity["필기응시자"].plot(kind="barh", grid=True, figsize=(10,10));
```

결과값을 봐도 모르겠다. 그 이유는 인덱스를 행번호로 설정했기 때문이다. 아래의 코드를 통해 인덱스를 변경하자

```python
Electricity.set_index("실시연도", inplace=True)
Electricity["필기응시자"].plot(kind="barh", grid=True, figsize=(10,10));
```

set_index는 기준을 변경한다는 이야기 이다. 원래 기준은 행번호 였는데, 이제 기준은 실시연도를 기준으로 하게된다. 


여기서 볼만한 정보는 응시자의 수는 2001년과 비교하면 약 6배 증가하였다.

```python
Electricity["실기합격자"].plot(kind="barh", grid=True, figsize=(10,10));
```

실기합격자는 2001년에 비해 약 6배 증가하였다.

```python
plt.figure(figsize=(20,10))
plt.xticks([2001, 2002, 2003, 2004, 2005, 2006,
              2007, 2008, 2009, 2010, 2011, 2012,
              2013, 2014, 2015, 2016, 2017, 2018,2019, 2020, 2021])
plt.plot(Electricity["필기응시자"], label = "필기응시자")
plt.plot(Electricity["실기합격자"], label = "실기합격자")
plt.grid()
plt.legend()
plt.xlabel("실시년도")
plt.show()
```

응시자는 점점 증가하는 추세이며 
전기기사에 도전하는 사람들(필기응시자)는 2001년에 비해 6배 증가하였는데 

실기합격자도 2001년에 비해 약 6배 증가하였다. 

```python
plt.figure(figsize=(20,5))
plt.xticks([2001, 2002, 2003, 2004, 2005, 2006,
              2007, 2008, 2009, 2010, 2011, 2012,
              2013, 2014, 2015, 2016, 2017, 2018,2019, 2020, 2021])
plt.plot(Electricity["최종합격률"], label = "실기합격률")
plt.grid()
plt.legend()
plt.xlabel("실시년도")
plt.show()
```

응시자가 증가하고, 관련 학원이나 교재들이 늘어나고 있는 상황이지만 합격률이 2010년도에 비해 높아졌다고 보기는 어렵다.

그렇기에 응시자가 늘어나고 관련학원이나 교재들이 늘어난다고 하여도 합격률이 높아지거나 하지는 않는다고 내 개인적으로 결론을 내릴 수 있다.
(하지만 이렇게 결론을 내릴 수는 없을 것이다. 관련학원이나 교재들의 수가 늘어났다는 근거자료가 있어야 하기 때문이다.

```python
plt.figure(figsize=(20,5))
plt.xticks([2001, 2002, 2003, 2004, 2005, 2006,
              2007, 2008, 2009, 2010, 2011, 2012,
              2013, 2014, 2015, 2016, 2017, 2018,2019, 2020, 2021])
plt.plot(Electricity["필기합격률"], label = "필기합격률")
plt.plot(Electricity["최종합격률"], label = "실기합격률")
plt.grid()
plt.legend()
plt.xlabel("실시년도")
plt.show()
```

전기기사를 공부하는 사람들 사이에서는 한가지 합리적의심이 있는데


"필기합격률이 높았단 해는 무조건 실기합격률이 낮다."

라는 이야기 이다.
