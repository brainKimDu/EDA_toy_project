```python
my_map = folium.Map(location=[37.5502, 126.982], zoom_start=11, title="Stamen Toner")

for each in range(len(ediya)):
    try:
        folium.Marker(
            location = [edi_y[each], edi_x[each]],
            popup = ediya["제목"][each],
            icon = folium.Icon(color = "blue")
        ).add_to(my_map)
            
    except:
        print("위도 경도값을 모르는 경우입니다.")
        
for each in range(len(starbucks)):
    try:
        folium.Marker(
            location = [star_y[each], star_x[each] ],
            popup = starbucks["제목"][each],
            icon = folium.Icon(color = "green")
        ).add_to(my_map)
    except:
        print("위도 경도값을 모르는 경우입니다.")       
    
```

## 스타벅스 주변에 이디야 매장이 몇 곳 있는가?
- 그런데 스타벅스의 몇 m 범위안에 있어야 이디야가 따라만든다고 할 수 있을까?


- 가산디지털단지에서 스타벅스간의 거리는 약 250m 이다.



```python
from haversine import haversine
```

스타벅스 매장 주변 250m 이내에 이디야 매장이 있는가를 탐색하는 코드

데이터처리 개수로 보면 비효율적이긴하다.

```python
Starbucks_nearby_ediya_250 = []
starbucks_gage = []
count = 0

for each in range(len(starbucks)):
        starbucks_gage1 = (float(star_y[each]), float(star_x[each]))
    
        ediya_gage = []
        for j in range(len(ediya)):
            ediya_gage1 = (float(edi_y[j]), float(edi_x[j]))
            ediya_gage.append(ediya_gage1)
            a = haversine(starbucks_gage1, ediya_gage1, unit = 'm')
            if a < 250:
                count = count + 1
        
        Starbucks_nearby_ediya_250.append(count)
        count = 0
Starbucks_nearby_ediya_250
```

```python
starbucks["주변 이디야"] = Starbucks_nearby_ediya_250
starbucks
```

```python
starbucks_ediya = starbucks
del starbucks_ediya["주소"]
del starbucks_ediya["구"]
del starbucks_ediya["경도"]
del starbucks_ediya["위도"]
starbucks_ediya
```

```python
count_ediya  = pd.DataFrame(starbucks_ediya.value_counts("주변 이디야"))
count_ediya = count_ediya.sort_values(by='주변 이디야' ,ascending=True)
count_ediya.rename(columns={count_ediya.columns[0]: "점포수"}, inplace=True)
count_ediya
```

```python
# matplotlib의 경우 한글폰트 깨짐현상을 방지하기 위한 코드
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

f_path = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

```python
f, ax = plt.figsize=(20,6)

count_ediya['점포수'].plot.pie(autopct = "%.1f")
plt.title('스타벅스 매장별 250m이내 이디야 매장의 수') 
plt.ylabel('')

plt.show()
```
