## 혹시 모르니 인덱스를 다시 한번 맞춰줍니다.

```python
starbuck_reindex = starbucks.reset_index()
starbuck_reindex.head(10)
```

이제 위도와 경도를 분리해내는 작업을 진행합니다.

우리가 0을 주거나 값이 이상한 경우가 있어서 그 경우에는 예외를 시켰습니다.

이디야의 위도 경도를 가져오는 과정입니다.

```python
x_data = []
y_data = []

for i in range(len(ediya["진위도/경도"])):
    temp = ediya["진위도/경도"]
    try:
        temp_ediya = temp[i].split(",")
        x = temp_ediya[0].split(":")
        x_value = x[1]
    except:
        print("예외")
        
    x_data.append(x_value)
    
    try:   
        temp_ediya = temp[i].split(",")
        y = temp_ediya[1].split(":")
        y_value = y[1]
    except:
        print("예외")
    
    y_data.append(y_value)
len(y_data)
```

```python
ediya["경도"] = x_data
ediya["위도"] = y_data
del ediya["진위도/경도"]
ediya
```

오류가 있다고 하지만 데이터는 이미 더 진행되어 있기 때문에 설명만 하고 넘어갑니다.
순서대로 진행하면 해결됩니다.

```python
x_data = []
y_data = []

for i in range(len(starbuck_reindex["진위도/경도"])):
    temp = starbuck_reindex["진위도/경도"]
    try:
        temp_star = temp[i].split(",")
        x = temp_star[0].split(":")
        x_value = x[1]
        x_data.append(x_value)
        
    except:
        x_data.append(0)
   
    
    try:
        temp_star = temp[i].split(",")
        y = temp_star[1].split(":")
        y_value = y[1]
        y_data.append(y_value)
        
    except: 
        y_data.append(0)
        

len(x_data)
```

```python
starbuck_reindex["경도"] = x_data
starbuck_reindex["위도"] = y_data
del starbuck_reindex["진위도/경도"]
starbuck_reindex
```

```python
del starbuck_reindex["index"]
starbuck_reindex
```

```python
ediya.to_csv(
    "ediya.csv", sep=',', encoding="UTF-8"
)

ediya.to_csv(
    "(2)ediya.csv", sep=',', encoding="UTF-8"
)

starbuck_reindex.to_csv(
    "starbucks.csv", sep=',', encoding="UTF-8"
)

starbuck_reindex.to_csv(
    "(2)starbucks.csv", sep=',', encoding="UTF-8"
)
```
